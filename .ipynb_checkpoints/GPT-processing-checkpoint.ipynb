```python
def run_gpt(system_prompt, query):
    done = False
    while not done:
        try:
            response = client.chat.completions.create(
            model="gpt-4-turbo-preview",
            messages=[
                {
                  "role": "system",
                  "content": system_prompt
                },
                {
                  "role": "user",
                  "content": query
                }
                ],
                temperature=0,
                max_tokens=2048,
                top_p=1,
                frequency_penalty=0,
                presence_penalty=0
            )
            done = True
        except Exception as e:
            # Handle exception
            print(f"GPT failed!, trying again The error is: {e}")
            time.sleep(10)

    return response
```

```python
def run_gpt_series(system_prompt, queries):
    responses = []
    messages = [{"role": "system", "content": system_prompt}]
    print("running query series")
    
    for query in queries:
        messages.append({"role": "user", "content": query})
        #print(query)
        retries = 0
        max_retries = 10
        while retries < max_retries:
            try:
                response = client.chat.completions.create(
                    model="gpt-4-turbo-preview",
                    messages=messages,
                    temperature=0,
                    max_tokens=2048,
                    top_p=1,
                    frequency_penalty=0,
                    presence_penalty=0
                )
                model_response = response.choices[0].message.content
                responses.append(model_response)

                messages.append({"role": "assistant", "content": model_response})
                break
            except Exception as e:
                print(f"GPT failed! The error is: {e}. Retrying...")
                time.sleep(10)
                retries += 1
            
        #print(model_response)
        
        if retries == max_retries:
            print("Max retries reached, moving to next query.")
    return responses
```

```python
def get_structured_contents(file_contents):
    #convert to struct of chapters, each divided into lesson, solved problems, and supplementary problems
    sections = re.split(r'(?=\\section..Chapter)', file_contents)
    contents = []
    for section in sections:
        print(section.split('\n', 1)[0],len(section))
        section_parts = {}
        section_parts["all"]=section.split('\n', 1)
        
        subsections = re.split(r'(?=\\section)', section)
        for subsection in subsections:
            if "solved problems" in subsection.split('\n', 1)[0].lower():
                print("-->"+subsection.split('\n', 1)[0])
            if "supplementary problems" in subsection.split('\n', 1)[0].lower():
                print("-->"+subsection.split('\n', 1)[0])
                
        # Splitting the section into three parts
        parts = section.split("Solved Problems", 1)
        part_one = parts[0]
        part_two_and_three = parts[1] if len(parts) > 1 else ""
        
        parts = part_two_and_three.split("Supplementary Problems", 1)
        part_two = "Solved Problems" + parts[0] if part_two_and_three else ""
        part_three_and_four = "Supplementary Problems" + parts[1] if len(parts) > 1 else ""

        #extract answers to supplementary problems
        parts = part_three_and_four.split("Answers to Supplementary Problems", 1)
        part_three = "Supplementary Problems" + parts[0] if part_two_and_three else ""
        part_four = "Answers to Supplementary Problems" + parts[1] if len(parts) > 1 else ""

        # Optionally, print or process the parts
        print("lesson:", len(part_one), "solved problems:", len(part_two), "supplementary:", len(part_three), ", total", len(part_one)+len(part_two)+ len(part_three))
        
        section_parts["lesson"]=part_one
        section_parts["solved_problems"]=part_two
        section_parts["supplementary_problems"]=part_three
        section_parts["answers_to_supplementary_problems"]=part_four
    
        if (len(part_one)>0 and len(part_two)>0):
            contents.append(section_parts)
        else:
            print("dropped a section from further analysis")

    return contents
```

```python
def get_solved_problems(contents):
    solved_problems_this_book = {}
    for chapter in range(0,len(contents)):
        response = run_gpt("You are a helpful assistant. You analyze latex documents, converting them into structured data. You always copy the relevant sections verbatim.  Output flat json with a simple list, do not use a key:value structure or anything more complex. For example: [1, 2, 3, \"four\", \"five\"]", 
            """Consider the following textbook in latex and list the "Solved Problem" numbers:
            """+contents[chapter]["solved_problems"]+"""
            List the "Solved Problems" numbers (number only)!""")
        
        content = response.choices[0].message.content
        print(content)
        
        #using json.loads would convert strings like 3.10 to 3.1, etc
        content_strings = content[1:-1].split(", ")
        
        for each in content_strings:
            print(each)

        system_prompt = "You are a helpful assistant. You analyze latex documents, extracting relevant sections. You always copy the relevant sections verbatim. Your ouptut is only the requested text, no other text or comments!"

        #convert to use run_gpt_series
        requests = []
        each = content_strings[0]
        requests.append("""Consider the following textbook in latex and copy verbatim the header and full contents of Solved problem """+str(each)+""":
    """+contents[chapter]["solved_problems"]+"""
    copy verbatim the header and full contents of Solved problem """+str(each)
        )
        
        for each in content_strings[1:]:
            requests.append("""ok. Now copy verbatim the header and full contents of Solved problem """+str(each))
            
        responses = run_gpt_series(system_prompt,requests)
        
        for i, response in enumerate(responses):
            print(each)
            print(response)
            solved_problems_this_book[content_strings[i]]=response
    
    return solved_problems_this_book
```

```python
def get_supplementary_problems(contents):
    supplementary_problems_this_book = {}
    #for each chapter
    for chapter in range(0,len(contents)):
        #first, get the list of supplementary questions IDs into content_strings
        response = run_gpt("You are a helpful assistant. You analyze latex documents, converting them into structured data. You always copy the relevant sections verbatim.  Output flat json with a simple list, do not use a key:value structure or anything more complex. For example: [1, 2, 3, \"four\", \"five\"]", 
            """Consider the following textbook in latex and list the "Supplementary Problem" numbers:
            """+contents[chapter]["supplementary_problems"]+"""
            List the "Supplementary Problems" numbers (number only)!""")
        
        content = response.choices[0].message.content
        print(content)
        
        #using json.loads would convert strings like 3.10 to 3.1, etc
        content_strings = content[1:-1].split(", ")

        #show off a bit
        for each in content_strings:
            print(each)

        #this is the general system prompt. It works fine
        system_prompt = "You are a helpful assistant. You analyze latex documents, extracting relevant sections. You always copy the relevant sections verbatim. Your ouptut is only the requested text, no other text or comments!"

        #Now extract the questions from (from this chapter)
        requests = []
        each = content_strings[0]
        #the first request contains the chapter text and the query
        requests.append("""Consider the following textbook in latex and copy verbatim the header and full contents of Supplementary problem """+str(each)+""":
        """+contents[chapter]["supplementary_problems"]+"""
        copy verbatim the header and full contents of Supplementary problem """+str(each)
        )

        #all following requests contain the query only!
        for each in content_strings[1:]:
            requests.append("""ok. Now copy verbatim the header and full contents of Supplementary problem """+str(each))
            
        responses = run_gpt_series(system_prompt,requests)

        #loop through responses, save them as questions into supplementary_problems_this_book
        for i, response in enumerate(responses):
            print(content_strings[i])
            print(response)
            
            supplementary_problems_this_book[content_strings[i]]={}
            supplementary_problems_this_book[content_strings[i]]["question"]=response

        #get the answers for questions in this chapter (this could be made more efficient by handling all answers at once but no let's not do that)
        requests = []
        each = content_strings[0]
        requests.append("""Consider the following list of worked answers and equations. Each of them corresponds to a question defined elsewhere, in the relevant chapter of the textbook. Copy the complete answer """+str(each)+""", including title ("""+str(each)+"""):
        """+contents[-1]["answers_to_supplementary_problems"]+"""
        Copy the contents of point """+str(each)+""", including title ("""+str(each)+"). If the point "+str(each)+" is missing, just write 'not contained in text'"
        )
        
        for each in content_strings[1:]:
            requests.append("""ok. Now copy the contents of point """+str(each)+""", including title ("""+str(each)+"). If the point "+str(each)+" is missing, just write 'not contained in text'")
            
        responses = run_gpt_series(system_prompt,requests)
        
        for i, response in enumerate(responses):
            print(content_strings[i])
            print(response)
            supplementary_problems_this_book[content_strings[i]]["answer"]=response
    
    return supplementary_problems_this_book
```
